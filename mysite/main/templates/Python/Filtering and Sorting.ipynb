```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv',delimiter='\t')
```

```python
chipo.head()
```

```python
chipo.loc[:,['item_name','item_price']].sort_values('item_name')
```

```python
chipo['item_price'] = chipo['item_price'].replace('[\$,]','',regex=True).astype(float)
chipo.loc[:,['quantity','item_price']].sort_values('item_price',ascending=False).head(1)
```

```python
veggie_bool = chipo['item_name'].str.match('Veggie Salad Bowl')
print(sum(veggie_bool))
```
