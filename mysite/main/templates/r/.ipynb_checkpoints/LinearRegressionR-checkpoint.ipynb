# Linear Regression
### Questions that Linear Regression can help to answer:
* is there a (linear) relationship between feature(s) (predictors) and the outcome variable (response)?
* How strong is the relationship?
* Which of the predictors are associated with the response?
* How accurately can we estimate the effect of a predictor on the reponse?
* How accurately can we predict outcomes outside of the data set after we have trained our model?
* Is there synergy among the predictors (is there an interaction)?

## Simple Linear Regression
Given a **single** predictor variable X and a response variable Y, we assume that there is a approximately a linear relationship between X and Y so that:

$ Y\approx \beta_{0} + \beta_{1} X+\epsilon $

$\beta_{0}$ and $\beta_{1}$ are unknown constants that are called **_intercept_** and **_slope_**, respectively. They are the **_coefficients_** or **_paramters_** of our model. $\epsilon$ represents the _irreducible error_ that has a mean of 0.

We use our dataset to estimate these coefficients to obtain our model:

$\hat{y}=\hat{\beta_{0}}+\hat{\beta_{1}}x$

Let's look at an example which consists of a dataset that has _Sales_ as a response variable and different advertising strategies (_TV_, _Radio_ and _Newspaper_) as features and we want to answer the above questions:

```r
setwd('~/Downloads')
data = read.csv('Advertising.csv')
head(data)
```

Our goal is to find our parameters $\beta_{0}$ and $\beta_{1}$ so that  the distance of the data points to resulting line is as short as possible. To do that, there are a several ways but the most common way is minimizing the **_least squares criterion_**. In the following graph we see that with increasing in expenditures on _TV_ advertising, the _Sales_ increase:

```r
plot(data$TV, data$Sales, xlab='TV', ylab='Sales')
tv_sales <- lm(data$Sales~data$TV)
abline(tv_sales)
```

The line represents what value for _Sales_ our linear model would predict if we give a certain value for _TV_. The distance between each predicted value and the true value is called **_residual_**: $e_{i}=y_{i}-\hat{y_{i}}$
And the sum of all squared distances is the **_Residual Sum of Squares (RSS)_**. The _least squares_ approach chooses the coefficients to minimize the RSS. In this case our coefficients are:

```r
tv_sales$coefficients
```

And thus our function is:  $\hat{f}(x)=7.033 + 0.048x$
Getting back to our question whether there exists a relationship between X and Y, we can perform a Hypothesis tests with

$H_{0}$: There is no relationship between X and Y or mathematically $\beta_{1}=0$  
$H_{1}$: There is a relationship between X and Y or mathematically $\beta_{1}\neq 0$

The following command returns the statistics we need:

```r
summary(tv_sales)
tv_sales_mse <- mean(residuals(tv_sales)^2)
tv_sales_rmse <- sqrt(tv_sales_mse)
cat('Mean Squared Error (MSE):', tv_sales_mse, '\n');
cat('Root Mean Squared Error (RMSE):', tv_sales_rmse )
```

Given the SE of 0.46 for the estimate $\hat{\beta_{1}}$ is relatively large, the t-value is large. The p-value shows how unlikely this is under $H_{0}$ and therefore we reject $H_{0}$.

## Multiple Linear Regression
The mulitple linear regression contains a slope coefficient for each predictor variable. In our example, we have 3 predicotr variables, so assuming a linear relationship, we get:
$ Y\approx \beta_{0} + \beta_{1} X_{1} + \beta_{2} X_{2} + \beta_{3} X_{3}+\epsilon $

and therefore our estimate is:
$\hat{y}=\hat{\beta_{0}}+\hat{\beta_{1}}x_{1} +\hat{\beta_{2}}x_{2} +\hat{\beta_{3}}x_{3}$

```r
tv_radio_np_sales <- lm(Sales~TV+Radio+Newspaper, data=data)
summary(tv_radio_np_sales)
```

We could have done a single linear regression for each predictor instead. See the following table:

```r
radio_sales <- lm(data$Sales~data$Radio)
np_sales <- lm(data$Sales~data$Newspaper)
test <- (c(coefficients(tv_radio_np_sales)[2],coefficients(tv_sales)[2],coefficients(tv_radio_np_sales)[3],coefficients(radio_sales)[2],coefficients(tv_radio_np_sales)[4],coefficients(np_sales)[2]))
lr_m <- matrix(c(coefficients(tv_radio_np_sales)[2],coefficients(tv_sales)[2],coefficients(tv_radio_np_sales)[3],coefficients(radio_sales)[2],coefficients(tv_radio_np_sales)[4],coefficients(np_sales)[2]), ncol = 2,byrow = T)
rownames(lr_m) <- c('TV','Radio','Newspaper')
colnames(lr_m) <- c('MLR','SLR')
round(lr_m,digits = 3)
```

We can see that the predictor _Newspaper_ was associated with _Sales_ in Single Linear Regression but not in Multiple Linear Regression. That is because _Newspaper_ and _Radio_ are correlated and the MLR shows that only _Radio_ does have influence on _Sales_.

We will now answer 5 questions with MLR:

1. Is at least one of the predictors useful in predicting the response?
Again we make a Hypothesis test:

$H_{0}$: There is no relationship between the predictor variables and Y or mathematically $\beta_{1} = \beta_{2} = \beta_{3} =0$  
$H_{1}$: There is a relationship between at least one of the predictors  and Y or mathematically $\beta_{j}\neq 0$

We perform this Hypothesis test using the F-statistic. Under the $H_{0}$ the F-statistic follows the F-distribution and we can calculate a p-value. In this case the likelihood that the F-Statistic is 570 under $H_{0}$ is <0.001.
